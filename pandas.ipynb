## Series
- A Pandas Series is like a column in a table.
- It is a one-dimensional array holding data of any type.

```python
cars=pd.Series(["BMW","Toyota","Honda"])
cars
```

## Labels
- labels are the identifiers used to access each element or value in the Series.
- The label serves as the index that helps you identify and retrieve specific values from the Series.
- If nothing else is specified, the values are labeled with their index number.
- The first value has index 0, the second value has index 1, etc.

```python
cust_index_cars=pd.Series(["BMW","Toyota","Honda"], index=["first","second","third"])
cust_index_cars
```

```python
cust_index_cars["first"]
```

```python
cust_index_cars["second"]
```

```python
cust_index_cars["third"]
```

```python
color_cars=pd.Series(["Red","Blue","White"])
color_cars
```

## Aggregation and Grouping

```python
df1.dtypes
```

```python
#group by college name and calculate the mean of cgpa
avg_cgpa=df1.groupby('college_name')['cgpa'].mean()
avg_cgpa
```

```python
#aggregation
df1['cgpa'].sum()
```

```python
df1['cgpa'].mean()
```

```python
#plot the highest cgpa based on the college
import matplotlib.pyplot as plt
highest_cgpa=df1.groupby('college_name')['cgpa'].max()
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='bar',color='skyblue')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('College Name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

#show the plot
plt.xticks(rotation=45)#rotate xaxis labels
plt.show()
```

```python
#pie chart
plt.figure(figsize=(7,7))
plt.pie(highest_cgpa,labels=highest_cgpa.index,autopct='%1.1f%%',startangle=140,colors=['skyblue','lightgreen','lightcoral'])
plt.title('percentage distribution of highest cgpa secured by college', fontsize=12)
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df1['stud_id'],df1['cgpa'],color='skyblue', edgecolor='black')

plt.title('scatter plot of student id vs cgpa',fontsize=14)
plt.xlabel('student id',fontsize=12)
plt.ylabel('cgpa',fontsize=12)

plt.show()
```
